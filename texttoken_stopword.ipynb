```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords

import spacy
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
text = "Machine learning and data science are very powerful technologies."
print(text)
```

```python
tokens = word_tokenize(text)
print("Word Tokens:", tokens)

sentences = sent_tokenize(text)
print("Sentence Tokens:", sentences)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("After Stopwords Removal:", filtered_tokens)
```

```python
! python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

spacy_tokens = [token.text for token in doc]
print("spaCy Tokens:", spacy_tokens)
```

```python
spacy_filtered = [token.text for token in doc if not token.is_stop]

print("spaCy After Stopwords Removal:", spacy_filtered)
```
